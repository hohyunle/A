```python
import math
def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

#알고리즘 5.8 띠 영역 내에서 d보다 작은 최근접 쌍의 거리 찾기
def strip_closest(P,d):
    n = len(P)                                      #리스트내의 점의 수
    d_min = d
    P.sort(key = lambda point : point[1])           #y축을 따라 정렬
    #merge(P, 0, len(P)-1)

    for i in range(n):                              #y가 최소인 점부터 순서대로
        j = i +1

        #P[i].y와 P[i].y의 차이가 d_min 이내일 때까지만 처리
        while j<n and (P[j][1]-P[i][1]) < d_min:
            dij = distance(P[i], P[j])
            if dij < d_min:
                d_min = dij

            j += 1
    return d_min

#알고리즘 5.9 최근접 쌍의 거리
def closest_pair_dist(P, n):
    if n <= 3:                              #점이 3개 이하이면, btute force로 바로 계산
        return closest_pair(P)              #억지 기법 알고리즘(알고리즘 3.4)
    
    mid = n //2                             #중앙점을 찾음. P는 현재 x로 정렬되어 있음
    mid_x = P[mid][0]                       #중앙점의 x좌표

    dl = closest_pair_dist(P[:mid], mid)    #Pl에서 dl 계산
    dr = closest_pair_dist(P[mid:], n-mid)  #Pr에서 dr 계산
    d = min(dl, dr)                         #d는 둘 중에서 더 짧은 거리

    Pm = []                                 #중앙에서 x좌표가 d 이내인 점들의 집합 Pm을 만듦

    for i in range(n):                      #Pm도 x에 대해 정렬되어 있음
        if abs(P[i][0] - mid_x) < d:
            Pm.append(P[i])

    ds = strip_closest(Pm, d)               #Pm내에서 d보다 작은 최근접쌍 거리 찾기

    return ds

def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])
            if dist < mindist:
                mindist = dist
    return mindist

def merge(A, left, mid, right):
    sorted = [0] * len(A)
    k = left
    i = left
    j = mid+1
    while i<= mid and j<=right:
        if A[i] <= A[j]:
            sorted[k] = A[i]
            i, k = i+1, k+1
        else:
            sorted[k] = A[j]
            j, k = j+1, k+1
            
    if i>mid:
        sorted[k:k+right-j+1] = A[j:right+1]

    else:
        sorted[k:k+mid-i+1] = A[i:mid+1]

    A[left:right+1] = sorted[left:right+1]


data = [(2,3), (12,30), (40,50),(5,1), (12,10), (3,4)]
data.sort(key = lambda point: point[0])

print("가장 가까운 두 점의 거리 : ", closest_pair_dist(data, len(data)))

```
