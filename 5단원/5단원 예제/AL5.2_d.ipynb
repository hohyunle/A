#### 1. 문제 정의 : 5.2 병합 알고리즘(정렬된 두 리스트의 병합)

#### 2. 알고리즘 설명
분할 정복 알고리즘으로써 하나의 리스트를 같은 크기의 두 개의 부분 리스트로 분할하고, 분할된 부분 리스트를 정렬한 다음, 정렬된 리스트를 병합하는 전략을 사용한다.

#### 3. 손으로 푼 예제


#### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요)
함수명 : merge(A, left, mid, right)         
입 력   변 수 : 정리할 리스트    
출력 : 정렬된 리스트 A   
함 수   설 명 : 주어진 리스트를 반으로 나누어 왼쪽 리스트, 오른쪽 리스트로 정렬한다. 이후 merge_sort 함수를 이용하여 나뉘고 정렬된 리스트를 하나로 합친다.

#### 5. 알고리즘 코드

```python
def merge(A, left, mid, right):
    k = left
    i = left
    j = mid + 1
    while i<=mid and j<=right:
        if A[i] <= A[j]:
            sorted[k] = A[i]
            i, k = i+1, k+1
        else:
            sorted[k] = A[j]
            j, k = j+1, k+1
    
    if i>mid:
        sorted[k:k+right-j+1] = A[j:right+1]
    else:
        sorted[k:k+mid-i+1] = A[i:mid+1]

    A[left:right+1] = sorted[left:right+1]
```
