## 1. 문제 정의

이진트리의 후위 순회

## 2. 알고리즘 설명

알고리즘은 이진 트리의 노드를 후위 순회하여 처리하는 방법을 제시한다. 후위 순회는 왼쪽 서브트리를 먼저 순회한 후에 오른쪽 서브트리를 순회하고 마지막으로 현재 노드를 처리하는 방식이다.

## 3. 손으로 푼 예제

## 4. 코드 개요

함수명: postorder

입력:
n (노드): 트리의 루트 노드 또는 서브트리의 루트 노드
출력:
없음

주어진 노드 n이 None이 아닌 경우에만 아래 단계를 수행한다.
왼쪽 서브트리를 후위 순회한다. 이는 postorder(n.left)를 재귀적으로 호출하여 왼쪽 서브트리를 후위 순회하는 것이다.
오른쪽 서브트리를 후위 순회한다. 이는 postorder(n.right)를 재귀적으로 호출하여 오른쪽 서브트리를 후위 순회하는 것이다.
현재 노드의 데이터를 처리한다. 이를 위해 print(n.data, end=' ')를 호출하여 현재 노드의 데이터를 출력한다.

## 5. 코드(py)

```python
def postorder(n) : 
    if n is not None :  # 만약 주어진 노드가 None이 아니라면
        postorder(n.left)  # 왼쪽 서브트리를 후위 순회합니다.
        postorder(n.right)  # 오른쪽 서브트리를 후위 순회합니다.
        print(n.data, end=' ')  # 현재 노드의 데이터를 출력하고 줄바꿈하지 않고 공백으로 처리합니다.
```

## 8. 복잡도 분석

알고리즘에서 출력문(print(n.data, end= '))을 기본 연산으로 하면 복잡도 함수는 명확히 T(n) = n 이고, O(n) 에 속한다.
