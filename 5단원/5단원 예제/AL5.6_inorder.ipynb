## 1. 문제 정의

이진트리의 중위 순회

## 2. 알고리즘 설명

이 알고리즘은 이진 트리의 노드를 중위 순회하여 처리하는 방법을 보여준다. 주어진 노드를 방문하고, 왼쪽 서브트리를 중위 순회한 후에 현재 노드를 처리하고, 마지막으로 오른쪽 서브트리를 중위 순회한다.

## 3. 손으로 푼 예제

## 4. 코드 개요


함수명: inorder

입력:
n (노드): 트리의 루트 노드 또는 서브트리의 루트 노드
변환값: 없음

만약 노드 n이 None이 아닌 경우:
왼쪽 서브트리를 중위 순회합니다 (재귀 호출: inorder(n.left)).
현재 노드의 데이터를 출력합니다 (print(n.data, end=' ')).
오른쪽 서브트리를 중위 순회합니다 (재귀 호출: inorder(n.right)).

## 5. 코드(py)

```python
def inorder(n) : # 중위 순회 함수 
    if n is not None : 
        inorder(n.left) # 왼쪽 서브 트리 처리 
        print(n.data, end=' ') # 루트노드 처리(화면 출력) 
        inorder(n.right) # 오른쪽 서브 트리 처리
```

## 6. 테스트

```python
def inorder(n) : # 중위 순회 함수 
    if n is not None : 
        inorder(n.left) # 왼쪽 서브 트리 처리 
        print(n.data, end=' ') # 루트노드 처리(화면 출력) 
        inorder(n.right) # 오른쪽 서브 트리 처리
class TNode: 
    def __init__ (self, data, left, right): 
        self.data = data 
        self.left = left 
        self. right= right

d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c =TNode('C', f, None)
root = TNode('A', b, c)
#루트 노드

print(' In-Ordeer : ', end='')
inorder(root)
print()
```

## 8. 복잡도 분석

출력문(print(n.data, end=' '))을 기본 연산으로 한다면, 중위 순회 알고리즘의 복잡도 함수는 명확히 T(n) = n이 되고, 따라서 O(n)에 속하게 된다。

그러나 만약 모든 알고리즘에서 2행의 공백 트리 검사를 기본 연산으로 한다면, 확장된 트리의 모든 노드에서 검사가 이루어져야 한다. 이 경우, 추가적인 공간적 비용이 들지만 시간 복잡도 측면에서는 여전히 T(n) = n이며, 따라서 여전히 O(n)에 속한다.

이렇게 함으로써 두 가지 경우 모두 알고리즘의 복잡도가 O(n)이다。
