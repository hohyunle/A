문제 설명: 2차원 평면상에 n개의 점이 있다. 가장 인접한 쌍의 거리를 구하라.

최종적으로 가장 작은 거리를 반환한다

## 4. 알고리즘 코드 개요

변수: 리스트 = p, 좌표 = x,y

함수 이름: 거리 계산(유클리드 거리) = distance, 최근접 쌍의 거리 = closest_pair

출력: mindist

수행순서: 

점의 리스트 p가 입력된다. [(x1,y1)....(xn,yn)]

일단 최근접 거리를 무한대로 한다.

i < j를 만족하는 모든 쌍 (pi,pj)에 대한 거리를 계산하고, mindist와 비교하여 이보다 작으면 mindist를 갱신한다.

최종적으로 mindist 에 최근접 쌍의 거리가 저장된다.

## 5. 알고리즘 코드(py)

```python
# math 모듈, 제곱근(sqrt) 함수 사용하기위해서
import math

# 유클리드 거리 계산 함수
def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

# 최근접 쌍의 거리
def closest_pair(p):
    n = len(p)                              #점의 전체 개수
    mindist = float("inf")                  #최근접 거리 초기화(무한대)
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])     #유클리드 거리 계산
            if dist< mindist:
               mindist = dist
    return mindist                          #출력
```

## 6. 테스트 코드(py)

```python
# math 모듈, 제곱근(sqrt) 함수 사용하기위해서
import math

# 유클리드 거리 계산 함수
def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

# 최근접 쌍의 거리
def closest_pair(p):
    n = len(p)                              #점의 전체 개수
    mindist = float("inf")                  #최근접 거리 초기화(무한대)
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])     #유클리드 거리 계산
            if dist< mindist:
               mindist = dist
    return mindist                          #출력
    
#입력
p = [(2,3), (12,30), (40,50), (5,1), (12,10),(3,4)]
#화면출력
print("최근접 거리: ", closest_pair(p))
```

## 8. 복잡도 분석

두 개의 반복문이 모두 n에 비례하는 횟수만큼 실행되기 때문입니다. 첫 번째 반복문이 n-1번 실행되고, 두 번째 반복문이 i+1부터 n까지 실행되기 때문에 대략 n(n-1)/2 번의 연산이 수행됩니다. 

따라서 전체적인 시간 복잡도는 O(n^2)가 됩니다.
