#### 2. 알고리즘 설명
먼저 테이블 크기를 입력받고 None으로 초기화를 한다
그리고 입력된 숫자들을 해시함수에 맞는 주소에 넣는데 중복된 값이 나올 경우 주소를 +1 하면서 빈 자리를 계속 찾는다.
마지막 자리까지 갔는데도 비어있는 곳이 없을 경우 첫번째 자리로 가서 다시 찾는다.

#### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요)
함수명 : lp_insert(key)

함 수   설 명 : 키값을 테이블에 삽입할 때 앞에 있는 키값부터 먼저 맞는 자리에 넣고 같은 자리값이 나온 키는 빈 자리가 나올 때 까지 한칸씩 +하면서 빈 자리를 찾는다

#### 5. 알고리즘 코드

```python
M = 13                              #테이블의 크기
table = [None]*M                    #테이블 만들기: None으로 초기화 
def hashFn(key) :                   #해시 함수
    return key % M

def lp_insert(key) :
    id = hashFn(key)
    count = M
    while count>0 and (table[id] != None) :
        id = (id +1 +M) & M         #다음 위치로 이동
        count -=1                   #검사할 남은 위치의 수
    if count > 0 :
        table[id] = key             #해당 슬롯에 항목 저장
    return
```
