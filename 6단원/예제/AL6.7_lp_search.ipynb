#### 1. 문제 정의 : 6.7 선형 조사법의 탐색 알고리즘

#### 2. 알고리즘 설명
삽입에서 입력한 값을 탐색에서 다시 입력 하여 저장된 키 값과 입력한 값이 맞을 때 까지 +1 하면서 비교를 한다 찾을 경우 성공하지만 다음 버킷이 비어있을 경우 실패한다.

#### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요)
함수명 : lp_search(key)      
입 력   변 수 : 찾고자 하는 값                  
출력 : 찾고자 하는 값이 있으면 그 값 반환  
함 수   설 명 : 탐색키가 입력되면 해시주소를 계산하고, 해당 주소에 같은 키의 레코드가 있다면 탐색은 성공한다. 

#### 5. 알고리즘 코드

```python
def lp_search(key) :
    id = hashFn(key)
    count = M
    while count>0 :
        if table[id] == None :  #찾는 항목이 테이블에 없음
            return None
        if table[id] == key :
            return table[id]    #찾는 항목이 테이블에 있으
        id = (id +1) % M        #없으면 다음 위치 검사
        count -=1
    return None
```
