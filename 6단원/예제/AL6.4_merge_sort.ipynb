#### 2. 알고리즘 설명

#### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요)
함수명 : search_horspool(T, P)      
입 력   변 수 : T = 텍스트, P = 패턴            
출력 : T에서 P가 나타나는 위치를 찾아 반환  
함 수   설 명 : 텍스트 문자열에서 패턴 문자열을 찾는 함수 


####   5. 코드


```python
def search_horspool(T, P):
    m = len(P)                      #패턴의 길이
    n = len(T)                      #텍스트(입력)의 길이
    t = shift_table(P)              #shift 테이블 생성
    i = m-1                         #패턴의 우측끝 문자 위치
    while(i <= n-1):                #갸능한 모든 위치에 대해
        k = 0                       #매칭된 개수
        while k <= m-1 and P[m-1-k]==T[i-k]: #뒤에서 앞으로
            k += 1                  #맞으면 계속 앞으로 진행
        if k == m :                 #매칭성고이 매칭위치(왼쪽) 반환
            return i-m+1        
        else :                      #매칭실패
            tc = t[ord(T[i-k])]     #T[i]의 테이블 참조 -> 건너뜀
            i += (tc-k)
    return -1                       #매칭실패 -1 반환
```

####  6. 테스트 코드 

```python
NO_OF_CHARS = 128
def shift_table(pat):
    m = len(pat)
    tbl = [m]*NO_OF_CHARS

    for i in range(m-1):
        tbl[ord(pat[i])] = m-1-i
    
    return tbl

def search_horspool(T, P):
    m = len(P)
    n = len(T)
    t = shift_table(P)
    i = m-1
    while(i <= n-1):
        k = 0
        while k <= m-1 and P[m-1-k]==T[i-k]:
            k += 1
        if k == m :
            return i-m+1
        else :
            tc = t[ord(T[i-k])]
            i += (tc-k)
    return -1

print("패턴의 위치 :",search_horspool("APPLEMANGOBANANAGRAPE", "BANANA"))
```
