#### 1. 문제 정의 : 7.1 피보나치수열(메모이제이션 이용)

#### 2. 알고리즘 설명
함수의 결과를 저장할 메모리 공간을 준비해 두고, 한번 계산한 값을 저장해 두었다가 재활용하는 최적화 기법을 **메모이제이션**이라 한다.

#### 3. 손으로 푼 예제


#### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요)
함수명 : fib_dp_mem(n)      
입 력   변 수 : 피보나치 정수 n       
출력 : 피보나치의 수    
함 수   설 명 : 먼저 풀리지 않은 문제인지 검사하고, 아직 풀리지 않았다면 순환식의 초기 조건을 검사해 해당하면 결과를 바로 계산한다. 만약 일반 상황이면 순환 호출한다. 또한 풀린 문제이거나, 풀었던 문제라면 그 결과는 모두 mem(n)에 저장한다.

#### 5. 알고리즘 코드

```python
def fib_dp_mem(n):
    if (mem[n]==None):
        if n<2:
            mem[n] = n
        else:
            mem[n] = fib_dp_mem(n-1) + fib_dp_mem(n-2)
    return mem[n]
```

#### 6. 테스트 코드

```python
def fib_dp_mem(n):
    if (mem[n]==None):
        if n<2:
            mem[n] = n
        else:
            mem[n] = fib_dp_mem(n-1) + fib_dp_mem(n-2)
    return mem[n]

n=8
mem = [None] * (n+1)
print('동적계획(메모이제이션): Fibonacci(%d) = '%n, fib_dp_mem(n))
```
