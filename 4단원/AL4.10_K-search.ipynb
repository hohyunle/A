#### 2. 알고리즘 설명
리스트에 있는 값을 순서대로 정렬한 후에, 찾는 값을 정렬한 값 순서대로 찾는 알고리즘입니다.

#### 3. 손으로 푼 예제


#### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요)
함 &nbsp;&nbsp;&nbsp;수&nbsp;&nbsp;&nbsp;명 : kth_smallest_sort(A, k)   
입 력   변 수 : 반복할 수 x와 반복할 횟수 n을 입력 받는다.  
출&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;력 : x**n을 거듭제곱한 결과값을 반환하며, 4.8 코드와 처리 시간을 계산해 출력한다.       
함 수   설 명 : 거듭제곱을 구하기 위한 함수로, 하향식 축소 정복기법을 사용합니다. 만약 반복할 횟수(n)이 0이면 1을 리턴하고, n이 짝수인 경우, power함수를 다시 호출해 x*x한 결과값을 x로, n//2한 값을 n에 입력해 호출한다. 또한 n이 홀수인 경우 두 번째 경우값에 *을 곱해 반복한다.

#### 5. 알고리즘 코드

```python
def kth_smallest_sort(A, k):
    A.sort()                #배열을 정렬
    return A[k-1]           #인덱스가 k-1인 항목 반환
```

#### 6. 테스트 코드

```python
#정렬을 이용한 k번째 작은 수 찾기
def kth_smallest_sort(A, k):
    A.sort()                #배열을 정렬
    return A[k-1]           #인덱스가 k-1인 항목 반환

A = [7,10,4,3,20,15]
print('A 출력 : ', A)

A.sort()                    #오름차순 정렬: A=[3,4,7,10,15,20]
print('오름차순 정렬 : ', A)

A.sort(reverse=True)        #내림차순 정렬: A=[20,15,10,7,4,3]
B=sorted(A, reverse=True)   #정렬된 리스트를 반환. A는 변경되지 않음

print('내림차순 정렬 : ', A)

array = [12,3,5,7,4,19,26,23,15]
print("입력 테스트= ", array)
print("[정렬기법] 3번째 작은 수: ", kth_smallest_sort(array, 3))
print("[정렬기법] 6번째 작은 수: ", kth_smallest_sort(array, 6))
```
