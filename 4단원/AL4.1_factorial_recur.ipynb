#### 1. 문제 정의 : 4.1 팩토리얼(하향식 축소 정복 기법)

#### 2. 알고리즘 설명
 "축소 정복" 기법 중 하나인 "하향식 접근"을 사용하여 팩토리얼을 계산하는 알고리즘입니다. 이 알고리즘은 재귀적인 구조를 이용하여 문제를 해결합니다.

#### 3. 손으로 푼 예제

#### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요)
함수명: factorial_recur   
축소 정복 기법: 하향식 접근 = 순환구조   
입력: n (양의 정수)   
변환값:  return (n * factorial_recur(n - 1)) 또는 1   

1. 입력으로 받은 정수 `n`이 1인지 확인합니다. 만약 1이라면, `n!`의 값은 1이므로 1을 반환합니다.   
2. 만약 `n`이 1이 아니라면, `n!`의 값을 구하기 위해 `n`을 `n-1`로 감소시킨 후, `factorial_recur(n-1)`을 호출하여 계산합니다. 그리고 `n`과 계산된 값을 곱한 결과를 반환합니다.

#### 5. 알고리즘 코드

```python
def factorial_recur(n) : # 축소 정복 기법: 하향식 접근 
    if n == 1 : 
        return 1 
    else : 
        return (n * factorial_recur(n - 1)) # 순환구조 이용
```

#### 6. 테스트 코드

```python
int count = 0
def factorial_recur(n) : # 축소 정복 기법: 하향식 접근 
    if n == 1 : 
        return 1 
    else : 
        count++
        return (n * factorial_recur(n - 1))
print(factorial_recur(10))

```

#### 8. 복잡도 분석
 재귀적인 구조로 인해 O(n)입니다. 이 함수는 입력값인 `n`에 대해 `n-1`씩 감소하면서 재귀 호출을 하기 때문입니다.
