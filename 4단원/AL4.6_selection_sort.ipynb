#### 1. 문제 정의 : 4.6 이진 탐색(반복구조)

#### 2. 알고리즘 설명 
이진 탐색 알고리즘을 반복구조로 기술 한 것으로 실행 효율성이 순환구조보다 좋다.

#### 3. 손으로 푼 예제 


#### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요)
입력 변수 :  [1, 3, 8, 13, 13, 16, 21, 26, 27, 30, 33, 36, 39, 41, 44, 49]  
출력 변수 :  10 or -1  
함수 설명 : 결과 10은 33이라는 숫자를 찾아냈을 때 표시됨 -1은 찾는 숫자가 없을 시 표시

#### 5. 알고리즘 코드


```python
def binary_search_iter(A, key, low, high):
    if(low <= high):
        mid = (low + high)
        if key == A[mid]:
            return mid
        elif key > A[mid]:
            low = mid + 1
        else :
            high = mid - 1
    return -1
```

#### 6. 테스트 코드 

```python
def binary_search_iter(A, key, low, high):
    while(low <= high):
        mid = (low + high)
        if key == A[mid]:
            return mid
        elif key > A[mid]:
            low = mid + 1
        else :
            high = mid - 1
    return -1

listA = [1, 3, 8, 13, 13, 16, 21, 26, 27, 30, 33, 36, 39, 41, 44, 49]
print("입력 리스트 = ", listA)
print("33 탐색(반복) -->", binary_search_iter(listA, 33, 0, len(listA)-1))
print("32 탐색(반복) -->", binary_search_iter(listA, 32, 0, len(listA)-1))
```
