#### 2. 알고리즘 설명
축소 정복 기법이란 주어진 문제를 해결하기 위해 이 문제와 좀 더 작은 문제간의 관계를 이용하는 전략을 말합니다. 이때 문제 사이의 관계가 구해지면, 하향식이나 상향식으로 적용할 수 있습니다. 하향식의 경우 순환구조를 사용하고, 상향식의 경우 반복구조를 사용합니다.

#### 4. 알고리즘 개요(입력 변수, 출력, 알고리즘 개요)
함 &nbsp;&nbsp;&nbsp;수&nbsp;&nbsp;&nbsp;명 : power    
입 력   변 수 : 반복할 수 x와 반복할 횟수 n을 입력 받는다.  
출&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;력 : x**n을 거듭제곱한 결과값을 반환하며, 4.8 코드와 처리 시간을 계산해 출력한다.       
함 수   설 명 : 거듭제곱을 구하기 위한 함수로, 하향식 축소 정복기법을 사용합니다. 모든 수의 0승은 1이기 때문에 값을 리턴하고, n이 짝수인 경우, power함수를 다시 호출해 x*x한 결과값을 x로, n//2한 값을 n에 입력해 호출한다. 또한 n이 홀수인 경우 두 번째 경우값에 *을 곱해 반복한다.

#### 5. 알고리즘 코드

```python
def power(x, n):
    if n == 0:
        return 1
    elif (n % 2) == 0:
        return power(x*x, n//2)
    else:
        return x * power(x*x, (n-1)//2)
```

#### 6. 테스트 코드

```python
def slow_power(x,n) :        #반복으로 구하는 함수
    result = 1.0
    for i in range(n):
        result = result * x  #루브 : n번 반복
    return result

def power(x, n):
    if n == 0:
        return 1
    elif (n % 2) == 0:
        return power(x*x, n//2)
    else:
        return x * power(x*x, (n-1)//2)
    
import time
print("억 지 기 법 (2**500)= ", power(2.0, 500))
print("축소정복기법(2**500)= ", slow_power(2.0, 500))

t1 = time.time()
for i in range(100000) : power(2.0, 500)
t2 = time.time()
for i in range(100000) : slow_power(2.0, 500)
t3 = time.time()

print("억 지 기 법  시간... ", t3-t2)
print("축소정복기법 시간... ", t2-t1)
```

#### 8. 복잡도 분석   
위 코드는 n이 짝수일 경우와 홀수일 경우로 나누어 연산을 처리합니다. 그러나 짝수일 경우 n을 절반으로 나누어 재귀적으로 호출하므로 O(log n)번의 재귀 호출이 이루어집니다. 또한 홀수일 경우엔 뺄셈 연산을 통해 n을 짝수로 만들고 결국 다시 짝수로 돌아가 재귀 호출을 합니다. 따라서 복잡도는 O(log n)임을 알 수 있습니다.
